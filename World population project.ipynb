# 2.WHAT DOES THE DISTRIBUTION OF POPULATION LOOK LIKE ACROSS DIFFERENT CONTINENT

```python
#PLOTTING THE POPULATION DISTRIBUTION ACROSS ALL CONTINENTS
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x = "Continent",y = "2020 Population",data = population)
plt.title("POPULATION DISTRIBUTION ACROSS ALL CONTINENTS")
plt.xlabel("Continent")
plt.ylabel("Population")
plt.show();
```

# 3. WHICH COUNTRIES HAVE THE HIGHEST AND LOWEST POPULATION IN 2020?

```python
#WE NEED TO FIND COUNTRIES WITH HIGHEST POPULATION.ATLEAST FIRST 5 COUNTRIES
```

```python
highest_population_2020 = population.nlargest(5,"2020 Population")
print("Countries with highest population in 2020:\n",highest_population_2020)
```

```python
#WE NEED TO KNOW COUNTRIES WITH LOWEST POPULATION IN 2020
```

```python
lowest_population_2020 = population.nsmallest(5,"2020 Population")
lowest_population_2020
print("Countries with lowest population in 2020 :\n",lowest_population_2020)
```

# 4.WHAT ARE THE TOP 10 MOST POPULOUS COUNTRIES IN 2020, AND  HOW DO THEIR POPULATION COMPARE

```python
#FIND THE TOP 10 MOST POPULOUS COUNTRIES IN 2020
```

```python
top_10_populous_countries = population.nlargest(10,"2020 Population")
top_10_populous_countries
```

```python
#TO PUT THIS IN A GRAPH FOR A BETTER VISUALIZATION
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x = "2020 Population",y = "Country/Territory",data = top_10_populous_countries)
plt.title("TOP 10 MOST POPULOUS COUNTRIES IN 2020")
plt.xlabel("POPULATION (in billions)")
plt.ylabel("COUNTRY/TERRITORY")
plt.show()
           
```

# 5.HOW DOES THE POPULATION GROWTH RATE VARY ACROSS THE CONTINENTS

```python
#FINDING THE POPULATION GROWTH RATE ACROSS THE CONTINENT
```

```python
population["Population_growth_rate"] = (population["2022 Population"]-population["2020 Population"])
```

```python
#PLOTTING POPULATION GROWTH RATE ACROSS CONTINENTS
plt.figure(figsize=(15,6))
sns.boxplot(x= "Continent",y = "Population_growth_rate",data = population)
plt.title("POPULATION GROWTH RATE ACROSS CONTINENT")
plt.xlabel("CONTINENT")
plt.ylabel("Population_growth_rate(%)")
plt.show()
```

# 6.WHICH ASIAN COUNTRIES HAVE A POPULATION OF MORE THAN 100 MILLION IN POPULATION.

```python
filtered_population_asia =population[( population["Continent"] == "Asia") & (population["2022 Population"] > 100_000_000)]
```

```python
#PLOTTING A GRAPH ON THIS
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x = "Country/Territory",y = "2022 Population",data = filtered_population_asia)
plt.title("ASIAN OVER POPULATED COUNTRIES")
plt.xlabel("COUNTRY/TERRITORY")
plt.ylabel("2022 POPULATION",fontsize = 15)
plt.xticks(rotation=45,ha = "right",fontsize = 12)
plt.yticks(fontsize = 12)
plt.tight_layout()
plt.show()
```
